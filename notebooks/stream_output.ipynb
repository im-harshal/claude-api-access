```python
# Load env variables
from dotenv import load_dotenv
load_dotenv()
```

```python
# Create an API client
from anthropic import Anthropic
client = Anthropic()
model = "claude-sonnet-4-5"
```

```python
# Helper function to add user and assistant messages to list and call claude api
def add_user_message(messages, text):
    user_message = { "role":"user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = { "role":"assistant", "content": text}
    messages.append(assistant_message)

def chat(messages, system=None, temperature=1.0):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature
    }

    if system:
        params["system"] = system

    message = client.messages.create(**params)
    return message.content[0].text
```

#### Temperature
Understanding temperature, below you see three calls made one after the other and claude model gave the same output as temperature was set to 0.0 but in the fourth call you see a different output as temperature was set to 1.0
close to 0.0 is more deterministic output
close to 1.0 is more random output 

```python
messages = []

add_user_message(messages, "Generate a short 1 sentence moive idea")

answer = chat(messages, temperature=0.0)
answer
```

#### Streaming

```python
messages = []

add_user_message(messages, "Write a 1 sentence description of a fake database")

stream = client.messages.create(
    model = model,
    max_tokens= 1000,
    messages = messages,
    stream = True
)

for event in stream:
    print(event)
```

```python
messages = []

add_user_message(messages, "Write a 1 sentence description of a fake database")

with client.messages.stream(
    model = model,
    max_tokens= 1000,
    messages = messages
) as stream:
    for text in stream.text_stream:
        print(text, end="")

stream.get_final_message()
```
